```python
import matplotlib.pyplot as plt

scenarios = [
    "No Exclusions",
    "Excluding Emails+Brands",
    "Excluding Categories+Brands",
    "Excluding Websites+Brands",
    "Excluding Addresses+Brands"
]

phi3_accuracy =     [90.4, 90.4, 92.8, 92.2, 92.2]
llama_accuracy =    [89.0, 68.2, 70.1, 68.2, 68.2]
electra_accuracy =  [85.7, 86.7, 76.2, 76.2, 76.2]

phi3_tp_fp =        [13.0, 13.0, 17.9, 14.1, 16.1]
llama_tp_fp =       [9.1,  2.6,  2.7,  2.6,  2.6]
electra_tp_fp =     [8.6,  8.0,  3.7,  3.7,  3.7]

phi3_tn_fn =        [6.53, 6.53, 9.06, 9.36, 8.26]
llama_tn_fn =       [6.74, 1.51, 1.73, 1.51, 1.51]
electra_tn_fn =     [4.03, 4.89, 2.5,  2.5,  2.5]

phi3_precision =    [0.928, 0.928, 0.947, 0.934, 0.942]
llama_precision =   [0.901, 0.721, 0.732, 0.721, 0.721]
electra_precision = [0.896, 0.889, 0.787, 0.787, 0.787]

phi3_recall =       [0.915, 0.915, 0.937, 0.940, 0.931]
llama_recall =      [0.922, 0.789, 0.810, 0.789, 0.789]
electra_recall =    [0.870, 0.895, 0.841, 0.841, 0.841]

phi3_f1 =           [0.922, 0.922, 0.942, 0.937, 0.936]
llama_f1 =          [0.912, 0.753, 0.769, 0.753, 0.753]
electra_f1 =        [0.883, 0.892, 0.813, 0.813, 0.813]

def plot_metric(title, phi, llama, electra, save_name=None):
    plt.figure(figsize=(8,5))
    plt.plot(scenarios, phi, marker='o', label="PHI-3")
    plt.plot(scenarios, llama, marker='o', label="Llama 3.2")
    plt.plot(scenarios, electra, marker='o', label="ELECTRA")
    plt.ylabel(title)
    plt.title(title)
    plt.grid(True)
    plt.legend()
    plt.xticks(rotation=25)
    plt.tight_layout()

    if save_name:
        plt.savefig(save_name, dpi=300, bbox_inches="tight")

    plt.show()


plot_metric("Accuracy",
            phi3_accuracy, llama_accuracy, electra_accuracy,
            save_name="accuracy.png")

plot_metric("True Positives:False Positives",
            phi3_tp_fp, llama_tp_fp, electra_tp_fp,
            save_name="tp_fp.png")

plot_metric("True Negatives:False Negatives",
            phi3_tn_fn, llama_tn_fn, electra_tn_fn,
            save_name="tn_fn.png")

plot_metric("Precision",
            phi3_precision, llama_precision, electra_precision,
            save_name="precision.png")

plot_metric("Recall",
            phi3_recall, llama_recall, electra_recall,
            save_name="recall.png")

plot_metric("F1 Score",
            phi3_f1, llama_f1, electra_f1,
            save_name="f1.png")
```
